```python
print(df.columns)
df["churn"]=df["churn"].astype("int")
df.dtypes
```

```python
x=np.asarray(df[['tenure', 'age', 'address', 'income', 'ed', 'employ', 'equip','internet']])
y=np.asarray(df["churn"])
x[0:5]

y[0:5]
```

```python
#normalize our datasets
from sklearn import preprocessing
x = preprocessing.StandardScaler().fit(x).transform(x)
x[0:5]
```

```python
#we will split our datasets
from sklearn.model_selection import train_test_split
X_train, X_test, y_train, y_test = train_test_split( x, y, test_size=0.2, random_state=4)
print ('Train set:', X_train.shape,  y_train.shape)
print ('Test set:', X_test.shape,  y_test.shape)
```

```python
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import confusion_matrix,jaccard_similarity_score,log_loss
LR = LogisticRegression(C=0.01, solver='liblinear').fit(X_train,y_train)
LR
```

```python
y=np.reshape(1,-1)
#Now we can predict using our test set:
pred=LR.predict(X_test)
pred
```

```python
yhat_prob = LR.predict_proba(X_test)
yhat_prob
```

```python
jaccard_similarity_score(y_test, pred)
```

```python
# Compute confusion matrix
cnf_matrix = confusion_matrix(y_test,pred, labels=[1,0])
np.set_printoptions(precision=2)
print(cnf_matrix)
```

```python
from sklearn.metrics import classification_report, confusion_matrix
print (classification_report(y_test, pred))
```
